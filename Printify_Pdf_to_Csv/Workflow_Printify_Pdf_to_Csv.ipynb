```python
def extract_invoice_data(pdf_path):
    # Open the PDF file
    with pdfplumber.open(pdf_path) as pdf:
        # Initialize variables to store extracted data
        net_amount_total = None
        shipping_cost = None
        tax_total = None
        invoice_amount = None
        
        # Iterate through each page in the PDF
        for page in pdf.pages:
            # Extract text from the page
            text = page.extract_text()
            
            # Search for keywords and extract data
            if "Net Amount Total (excl tax)" in text:
                net_amount_total = text.split("Net Amount Total (excl tax)")[1].strip()
            if "Shipping cost" in text:
                shipping_cost = text.split("Shipping cost")[1].strip()
            if "Tax" in text:
                tax_total = text.split("Tax Total")[1].strip()
            if "Invoice Amount" in text:
                invoice_amount = text.split("Invoice Amount")[1].strip()

        # Return the extracted data as a dictionary
        return {
            "Net Amount Total (excl tax)": net_amount_total,
            "Shipping cost": shipping_cost,
            "Tax Total": tax_total,
            "Total Invoice Amount": invoice_amount
        }

# Path to your PDF file
pdf_path = "C:\\Users\\joned\\Desktop\\Printify_June_2023_Invoice.pdf"

# Extract data from the PDF
invoice_data = extract_invoice_data(pdf_path)

# Check if the data was found
if all(value is not None for value in invoice_data.values()):
    # Define the CSV file path
    csv_file = "invoice_data_June.csv"
    
    # Write the data to a CSV file
    with open(csv_file, 'w', newline='') as csvfile:
        csvwriter = csv.writer(csvfile)
        csvwriter.writerow(["Description", "Value"])
        for key, value in invoice_data.items():
            csvwriter.writerow([key, value])

    print("Data extracted and saved to invoice_data.csv")
else:
    print("One or more values not found in the PDF.")
```

```python
import pdfplumber
import re
import pandas as pd

def extract_table_from_pdf(pdf_path):
    with pdfplumber.open(pdf_path) as pdf:
        all_text = ""
        for page in pdf.pages:
            all_text += page.extract_text()

        # Specific regex pattern with the re.DOTALL flag
        pattern = r"Net Amount Total.*?\(excl tax\)\s*USD ([\d.]+)\s*USD ([\d.]+)\s*USD ([\d.]+)\s*USD ([\d.]+)"

        matches = re.search(pattern, all_text, re.DOTALL)

        if matches:
            net_amount = matches.group(1)
            shipping_cost = matches.group(2)
            tax = matches.group(3)
            total_invoice_amount = matches.group(4)
        else:
            net_amount, shipping_cost, tax, total_invoice_amount = "Not Matched", "Not Matched", "Not Matched", "Not Matched"

        print("Net Amount Total (excl tax):", net_amount)
        print("Shipping cost:", shipping_cost)
        print("Tax:", tax)
        print("Total invoice amount:", total_invoice_amount)

        # Create DataFrame and save
        df = pd.DataFrame({
            "Net Amount Total (excl tax)": [net_amount],
            "Shipping cost": [shipping_cost],
            "Tax": [tax],
            "Total invoice amount": [total_invoice_amount]
        })
        
        return df

if __name__ == "__main__":
    pdf_path = "C:\\Users\\joned\\Desktop\\Printify_June_2023_Invoice.pdf"
    df = extract_table_from_pdf(pdf_path)
    df.to_csv('Printify_June_2023_Invoice.csv', index=False)
```
